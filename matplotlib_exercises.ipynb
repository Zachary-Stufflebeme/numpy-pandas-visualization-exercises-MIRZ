```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

#### Use matplotlib to plot the following equation:

y = (x ** 2) − x + 2.
You'll need to write the code that generates the x and y points.

Add an anotation for the point 0, 0, the origin.

```python

# Creating vectors X and Y
x = np.linspace(-50, 50, 100)
y = (x ** 2) - x + 2

# Create the plot
plt.plot(x, y)
plt.text(-5, -100, 'Origin', fontsize=10, color='blue')
# Show the plot
plt.show()
```

Create and label 4 separate charts for the following equations (choose a range for x that makes sense):

|y = √x|------------|y = x ** 3|--------------|y = 2 ** x|------------------|y = 1 / ( x + 1)|
You can use functions from the math module to help implement some of the equations above.

```python
# y = √x
x = np.linspace(0,100,100)
y = np.sqrt(x)
plt.plot(x,y)
plt.show()
```

```python
# y = x ** 3
x = np.linspace(0,10)
y = x ** 3

plt.plot(x,y)
plt.show()
```

```python
# y = 2 ** x
x = np.linspace(0,10)
y = 2 ** x
plt.plot(x,y)
plt.show()
```

```python
x = np.linspace(0,50)
y = 1 / ( x + 1)
plt.plot(x,y)
plt.show()
```

#### Combine the figures you created in the last step into one large figure with 4 subplots

```python
plt.subplot(2, 2, 1)
x = np.linspace(0,100)
y = np.sqrt(x)
plt.plot(x,y)
plt.title('y = √x')
plt.subplot(2,2,2)
x = np.linspace(0,10)
y = x ** 3
plt.plot(x,y)
plt.title('y = x ** 3')
plt.subplot(2,2,3)
x = np.linspace(0,10)
y = 2 ** x
plt.plot(x,y)
plt.title('y = 2 ** x')
plt.subplot(2,2,4)
x = np.linspace(0,50)
y = 1 / ( x + 1)
plt.plot(x,y)
plt.title('y = 1 / ( x + 1)')
plt.tight_layout()
plt.show()
```

#### Combine the figures you created in the last step into one figure where each of the 4 equations has a different color for the points. Be sure to include a legend and an appropriate title for the figure.

```python
plt.plot(np.linspace(0,100), np.sqrt(x), label = '√x')
plt.plot(np.linspace(0,100), x ** 3, label = ' x ** 3')
plt.plot(np.linspace(0,100), 2 ** x, label = '2 ** x')
plt.plot(np.linspace(0,100), (1 / ( x + 1)), label = '1 / ( x + 1)')
plt.xlim([0,100])
plt.ylim([0,100])
plt.legend()
plt.show()
```
