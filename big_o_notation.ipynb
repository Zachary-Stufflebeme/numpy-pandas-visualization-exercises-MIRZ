```python
n = np.linspace(0,100)
y = 1
my_list = ['1'] * 50
```

```python
plt.title('Big O Notation')
plt.xlabel('Elements')
plt.ylabel('Operations')
plt.plot(np.linspace(0,100), my_list, label = 'O(1)')
plt.plot(np.linspace(0,100),(np.log(n)), label = 'O(log(n))')
plt.plot(np.linspace(0,100),(n), label = 'O(n)')
plt.plot(np.linspace(0,100),(n * (np.log(n))), label = 'O(n log n)')
plt.plot(np.linspace(0,100),(n ** 2), label = 'O(n^2)')
plt.plot(np.linspace(0,100),(2 ** n), label = 'O(2^n)')
plt.plot(np.linspace(0,100), gamma(n) , label = 'O(n!)')
plt.plot(np.linspace(0,100),(n ** n), label = 'O(n^n)')
plt.ylim(0,1000)
plt.legend()
plt.show()
```
