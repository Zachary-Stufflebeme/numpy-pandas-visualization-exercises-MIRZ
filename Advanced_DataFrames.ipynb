#### 1. Use your get_db_url function to help you explore the data from the chipotle database.

```python
chipotle = my_db('''Select * From orders''', 'chipotle')
chipotle
```

#### 2. What is the total price for each order?

```python
chipotle['item_price'] = chipotle['item_price'].str.strip('$')
```

```python
chipotle['item_price'] = pd.to_numeric(chipotle.item_price)
```

```python
chipotle['order_total'] = chipotle.order_id
for x in np.unique(chipotle.order_id.values):
    chipotle.order_total[chipotle.order_id == x] = sum(chipotle.item_price[chipotle.order_id == x])
chipotle
```

#### 3. What are the most popular 3 items?

```python
top_three = chipotle.groupby('item_name').quantity.sum().sort_values(ascending=False).head(3)
top_three
```

#### 4. Which item has produced the most revenue?

```python
(chipotle.groupby('item_name').quantity.sum().sort_values() * chipotle.groupby('item_name').item_price.sum()).sort_values(ascending = False).head()
```
